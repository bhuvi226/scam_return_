```python
import pandas as pd

# Load your dataset
df = pd.read_csv('scam.csv')

# Update the column names to match your dataset
df = df[['return_reason', 'label']]  # Select the correct columns
df.columns = ['text', 'label']  # Rename columns to 'text' and 'label' for consistency

# Check the first few rows to confirm the changes
print(df.head())
```

```python
# You can perform any preprocessing here (like removing special characters, etc.)
df['text'] = df['text'].str.replace('[^a-zA-Z\s]', '', regex=True)  # Remove non-alphabetic characters

# Example: drop rows with missing values (if any)
df = df.dropna()

# Check the cleaned data
print(df.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score

# Upload the scam.csv file
#from google.colab import files
#uploaded = files.upload()

# Load the dataset from the uploaded file
df = pd.read_csv('scam.csv')

# Check the columns in the dataset to ensure it's correct
print(df.columns)  # Check columns to verify

# Assuming 'return_reason' is the text and 'label' is the target label
df['return_reason'] = df['return_reason'].fillna("")  # Handling any missing text data
df['label'] = df['label'].fillna(0)  # Handle missing labels (assuming 0 for non-scam)

# Vectorize the text using TF-IDF
vectorizer = TfidfVectorizer(max_features=800)  # Limit to 5000 features for simplicity
X = vectorizer.fit_transform(df['return_reason']).toarray()  # Transform text data into numerical features

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, df['label'], test_size=0.2, random_state=42)

# Build a simple neural network model
model = Sequential([
    Dense(64, activation='relu', input_dim=X_train.shape[1]),
    Dense(1, activation='sigmoid')  # Binary classification output
])

# Compile the model
model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=2)

# Evaluate the model on the test data
y_pred = (model.predict(X_test) > 0.5)  # Convert probabilities to binary labels

# Calculate and print the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
# Save the model
model.save('scam_detector.h5')

# Save the vectorizer
import pickle
with open('vectorizer.pkl', 'wb') as file:
    pickle.dump(vectorizer, file)
```
